```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
housing_data = pd.read_csv("housing.csv")
```

```python
housing_data
```

```python
housing_data.info() # here non null values are not similar so be got to drop off the null values using dropna
```

```python
housing_data.dropna(inplace = True) #for permanent changes
```

```python
housing_data.info() 
```

```python
#notice how the non null values are similar now
#now we are dividing our data for training and testing 
from sklearn.model_selection import train_test_split
```

```python
x = housing_data.drop(['median_house_value'],axis =1 ) #independent
y = housing_data['median_house_value'] #dependent 
```

```python
x_train , x_test , y_train, y_test = train_test_split(x,y,test_size = 0.2)#20% is testing 
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data['total_rooms']  = np.log(train_data['total_rooms']+ 1)
train_data['total_bedrooms']  = np.log(train_data['total_bedrooms']+ 1)
train_data['population']  = np.log(train_data['population']+ 1)
train_data['households']  = np.log(train_data['households']+ 1)#since some of the data is skewed 
```

```python
train_data.hist(figsize = (15,8))
```

```python
#i'm assuming here the ocean view can also affect the housing prices, like houses near to ocean are more expensive compared to others etc...
train_data =train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'],axis = 1)
```

```python
#houses with <1 hr  from ocean and > 1 hr from ocean represented by 1's and 0's respectively
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="latitude",y= "longitude", data= train_data,hue = "median_house_value",palette= 'coolwarm')
```

```python
#red is more expensive , blue is less expensive house
```

```python
# as you can see all the houses closer to ocean view are more expensive considering longitude, latitude
train_data['bedroom_ratio'] = train_data['total_bedrooms']/train_data['total_rooms']
```

```python
train_data['household_rooms'] = train_data['total_rooms']/train_data['households']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True,cmap="YlGnBu")
```

```python
from sklearn.linear_model import LinearRegression
x_train , y_train = train_data.drop(['median_house_value'],axis=1),train_data['median_house_value']
regression = LinearRegression()
regression.fit(x_train,y_train)
```

```python
test_data  = x_test.join(y_test)

test_data['total_rooms']  = np.log(test_data['total_rooms']+ 1)
test_data['total_bedrooms']  = np.log(test_data['total_bedrooms']+ 1)
test_data['population']  = np.log(test_data['population']+ 1)
test_data['households']  = np.log(test_data['households']+ 1)
test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)
test_data['bedroom_ratio'] = test_data['total_bedrooms']/test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms']/test_data['households']

```

```python
x_test, y_test = test_data.drop(['median_house_value'],axis = 1), test_data['median_house_value']
```

```python
regression.score(x_test,y_test)
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x=y_test, y=regression.predict(x_test), scatter_kws={'s': 20, 'alpha': 0.5}, line_kws={'color': 'red'})
plt.title('Linear Regression: Actual vs Predicted values with Regression Line')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor()
forest.fit(x_test,y_test)
```

```python
forest.score(x_test,y_test)
```

```python
# Example values for prediction
new_data = {
    'longitude': -120.0,
    'latitude': 35.0,
    'housing_median_age': 25.0,
    'total_rooms': 3000.0,
    'total_bedrooms': 600.0,
    'population': 1500.0,
    'households': 550.0,
    'median_income': 4.0,
    '<1H OCEAN': 0,
    'INLAND': 1,
    'ISLAND': 0,
    'NEAR BAY': 0,
    'NEAR OCEAN': 0,
    'bedroom_ratio': 0.2,
    'household_rooms': 5.5
}
new_data_df = pd.DataFrame([new_data])


new_data_df['total_rooms'] = np.log(new_data_df['total_rooms'] + 1)
new_data_df['total_bedrooms'] = np.log(new_data_df['total_bedrooms'] + 1)
new_data_df['population'] = np.log(new_data_df['population'] + 1)
new_data_df['households'] = np.log(new_data_df['households'] + 1)

prediction = forest.predict(new_data_df)

print("Predicted Median House Value:", prediction)
```
